ABSTRACT: This study explores the application of the Transformers architecture, specifically BERT-like models, in tasks beyond its usual domain of classification - focusing on numerical regression and quantized classification. The primary objective is to assess the performance of these models in predicting vowel density in text data and categorizing them into discrete classes. We utilize a corpus from B2W reviews, focusing on the 'review text' column, and adapt the BERTimbau model for our tasks.

# Comparing Task 2 and Task 3

```python
import pandas as pd

# Carregar os DataFrames
task2 = pd.read_csv('/content/drive/MyDrive/EP02_Regression_BERT/evaluation_metrics_classification.csv')
task3 = pd.read_csv('/content/drive/MyDrive/EP02_Regression_BERT/evaluation_metrics_classification_task3.csv')

# Adicionar uma coluna identificando a tarefa
task2['Task'] = 'Task 2'
task3['Task'] = 'Task 3'

# Concatenar os DataFrames
compare = pd.concat([task2, task3], ignore_index=True)

colunas = ['Task'] + [col for col in compare if col != 'Task']
compare = compare[colunas]

compare_transposed = compare.transpose()

# Exibir o DataFrame transposto
display(compare_transposed)
```
